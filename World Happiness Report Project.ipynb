```python
for i in df.columns:
    print(df[i].value_counts())
    print("\n")
```

```python
# Seperating numerical and categorical columns
# Checking for categorical column
cat_col=[]
for i in df.dtypes.index:
    if df.dtypes[i]=="object":
        cat_col.append(i)
print("Categorical Columns are:",cat_col)
print("\n")

# Checking for numerical column
num_col=[]
for i in df.dtypes.index:
    if df.dtypes[i]!="object":
        num_col.append(i)
print("Numerical Columns are:",num_col)
print("\n")
```

```python
# Checking nmber of unique values in each column
df.nunique().to_frame("No. of Unique values")
```

```python
# Statistical summary of numerical columns
df.describe()
```

1. The counts of all the columns are same which means there are no missing values in the dataset
2. The mean value is greater than the median(50%) in Happiness Score, Standard Error, Trust, Generosity and Dystopia Residual,  which means the data is skewed to right in these columns
3. The mean value is lesser than the median(50%) in Economy, Family, Health and Freedom,  which means the data is skewed to left in these columns
4. By summerizing the data we can observe that there is a very less difference between 75% and max, hence there are no outliers present in the data

```python
# Visualize the number of regions
ax=sns.countplot(x='Region',data=df)
print(df['Region'].value_counts())
```

```python
plt.figure(figsize=(15,20),facecolor="white")
plotnumber=1
for col in num_col:
    if plotnumber<=10:
        ax=plt.subplot(5,2,plotnumber)
        sns.distplot(df[col],color="m")
        plt.xlabel(col, fontsize=12)
        plt.yticks(rotation=0,fontsize=10)
    plotnumber+=1
plt.tight_layout()
```

```python
# Comparing Standard Error and Economy
plt.title("Comparison between Standard Error and Economy")
sns.scatterplot(x='Family',y='Generosity', data=df, hue='Happiness Score', palette='dark')
plt.show()
```
